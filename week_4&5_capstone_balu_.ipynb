````
!pip install geocoder
import geocoder
#finding the coordinates of the area names or neighbourhoods in london.
def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, London, England'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
````

````
#Now, let's get the top 100 venues that are within a radius of 2000 meters.#
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(area_name['Latitude'], area_name['Longitude'], area_name['Area_name']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
````

````
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['Area_name', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df.head()
````

````
# check if the results contain "Shopping Mall"
"Bar" in venues_df['VenueCategory'].unique()
````

````
# one hot encoding
kl_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
kl_onehot['Area_name'] = venues_df['Area_name'] 

# move neighborhood column to the first column
fixed_columns = [kl_onehot.columns[-1]] + list(kl_onehot.columns[:-1])
kl_onehot = kl_onehot[fixed_columns]

print(kl_onehot.shape)
kl_onehot.head()
````

````
#Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category


kl_grouped = kl_onehot.groupby(["Area_name"]).mean().reset_index()

print(kl_grouped.shape)
kl_grouped
````

````
len(kl_grouped[kl_grouped["Bar"] > 0])
````

````
lon_bar = kl_grouped[["Area_name","Bar"]]
lon_bar.head(37)
````

````
# set number of clusters
kclusters = 3

kl_clustering = lon_bar.drop(["Area_name"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kl_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
````

````
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
kl_merged = lon_bar.copy()

# add clustering labels
kl_merged["Cluster Labels"] = kmeans.labels_
````

````
kl_merged.rename(columns={"Area_name": "Neighborhood"}, inplace=True)
kl_merged.head()
````

````
kl_merged = kl_merged.join(area_name.set_index('Area_name'), on="Neighborhood")

print(kl_merged.shape)
kl_merged.head() # check the last columns!
````

````
kl_merged.sort_values(["Cluster Labels"], inplace=True)
kl_merged
````
